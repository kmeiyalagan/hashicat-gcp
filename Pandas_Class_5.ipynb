```python
data1 = {'key':['K0','K1','K2','K3'],
         'Key1':['K0','K1','K0','K1'],
         'Name':['Jai','Piyush','Gaurav','Anuj'],
         'Age':[27,24,22,32],}

data2 = {'key':['K0','K1','K2','K3'],
         'Key1':['K0','K0','K0','K0'],
         'Address':['Nagpur','Kanpur','Allahbad','Kannuaj'],
         'Qualification':['Btech','BA','BCom','B.Hons']}
```

```python
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
```

```python
df
```

```python
df1
```

```python
# LEFT
res = pd.merge(df,df1,how = 'left',on = ['key','Key1'])
res
```

```python
# RIGHT
res = pd.merge(df,df1,how = 'right',on = ['key','Key1'])
res
```

```python
# OUTER
res = pd.merge(df,df1,how = 'outer',on = ['key','Key1'])
res
```

```python
# INNER
res = pd.merge(df,df1,how = 'inner',on = ['key','Key1'])
res
```

```python
rng = pd.DataFrame()
rng['Date'] = pd.date_range('1/1/2022',periods = 72,freq = 'H')
rng
```

```python
rng['year'] = rng['Date'].dt.year
rng['month'] = rng['Date'].dt.month
rng['day'] = rng['Date'].dt.day
rng['hour'] = rng['Date'].dt.hour
rng['minute'] = rng['Date'].dt.minute
rng.head(10)
```
