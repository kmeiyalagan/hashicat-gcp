# Comcast Telecom Consumer Complaints Analysis

````python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv('/content/Comcast_telecom_complaints_data.csv')
````

**Data Wrangling / transformation**

````python
data["Count"] = 1
data["date_index"] = data["Date_month_year"] + " " + data["Time"]
data["date_index"] = pd.to_datetime(data["date_index"])
data["Date_month_year"] = pd.to_datetime(data["Date_month_year"])
````

````python
data = data.set_index(data["date_index"])
````

**1. trend chart for the number of complaints at monthly and daily**

````python
Month_view = data.pivot_table(index='Date_month_year', values='Count',aggfunc='sum',sort=True)
````

````python
fig, axs = plt.subplots(figsize=(12, 6))
Month_view.plot.area(ax=axs)
axs.set_ylabel("No of Compliants")
````

````python
daily_view = data.pivot_table(index='Date_month_year', values='Count',aggfunc='sum',sort=True)
````

````python
fig, axs = plt.subplots(figsize=(20, 6))
daily_view.plot.area(ax=axs)
axs.set_ylabel("No of Compliants")
````

**2. frequency of complaint types**

````python
Frequency = data.pivot_table(index='Received Via', values='Count',aggfunc='sum',sort=True)
Frequency
````
